```python
from slava.modules.data_loader import DataLoader
```

```python
data_loader = DataLoader()
dataset = data_loader.load_data()
```

```python
dataset
```

```python
import pandas as pd
import json

data = []
with open('answers.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        entry = json.loads(line)
        data.append(entry['answer'])

df = pd.DataFrame(data, columns=['response'])
```

```python
df
```

```python
df['model'] = 'evaluated_example'
df['Вид вопроса'] = [meta['type'] for meta in dataset['meta']]
df['Предмет'] = [meta['subject'] for meta in dataset['meta']]
df['Провокационность оценка LLM 3 класса'] = [meta['provac_score'] for meta in dataset['meta']]
```

```python
df
```

```python
df['Ответ'] = dataset['outputs']
```
