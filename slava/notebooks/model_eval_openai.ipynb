```python
import logging
from openai import OpenAI

class ModelHandler:
    """A class to handle model loading and response generation using OpenAI API."""

    def __init__(self, api_key: str):
        """
        Initializes the ModelHandler with the provided API key.

        Args:
            api_key (str): Your OpenAI API key.
        """
        self.api_key = api_key
        self.client = OpenAI(api_key=api_key)

    def generate_response(self, prompt: str) -> str:
        """Generates a response from OpenAI API based on the given prompt.

        Args:
            prompt (str): The prompt to send to the model.

        Returns:
            str: The model's response.
        """
        try:
            completion = self.client.chat.completions.create(
                model="gpt-4o",  
                messages=[
                    {"role": "system", "content": "You are a helpful assistant."},
                    {"role": "user", "content": prompt}
                ],
                max_tokens=150,
                temperature=0.7
            )
            
            return completion.choices[0].message.content.strip()
        except Exception as e:
            logging.error(f"Error while generating response: {e}")
            return "Ошибка при генерации ответа."


```
